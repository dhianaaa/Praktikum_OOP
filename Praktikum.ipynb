```python
def check_even_odd(n): return "The number is Even" if n % 2 == 0 else "The number is Odd"
print(check_even_odd(4)) 
print(check_even_odd(7))
```

```python
def check_number(n): 
    return "The number is Positive" if n > 0 else "The number is Negative" if n < 0 else "The number is Zero"
print(check_number(10)) 
print(check_number(-5))  
print(check_number(0))
```

```python
def is_anagram(str1, str2): return sorted(str1) == sorted(str2)
print(is_anagram("listen", "silent"))
print(is_anagram("hello", "world"))
```

```python
def factorial(n): return 1 if n == 0 else n * factorial(n-1)
print(factorial(5))
print(factorial(0))
```

```python
def is_palindrome(s): return s == s[::-1]
print(is_palindrome("racecar"))
print(is_palindrome("python"))
print(is_palindrome("habibah"))
```

```python
def is_armstrong(n): return n == sum(int(d)**len(str(n)) for d in str(n))
print(is_armstrong(153))
print(is_armstrong(370))
print(is_armstrong(123))
```

```python
class BankAccount:
    def __init__(self, name):
        self.name = name
        self.balance = 0
    
    def deposit(self, amount):
        self.balance += amount
        return f"Deposited {amount}. New balance: {self.balance}"
    
    def withdraw(self, amount):
        if amount > self.balance or amount <= 0:
            return "Invalid withdrawal amount or insufficient funds"
        self.balance -= amount
        return f"Withdrew {amount}. New balance: {self.balance}"

account = BankAccount("Name")
print(account.deposit(1000))
print(account.withdraw(500))
print(account.withdraw(600))
```

```python
class Student:
    def __init__(self, name):
        self.name = name
        self.grades = []
    
    def add_grade(self, grade):
        self.grades.append(grade)
        return f"Grade {grade} added."
    
    def get_average(self):
        avg = sum(self.grades) / len(self.grades)
        return f"Average grade: {avg}"

student = Student("Alice")
print(student.add_grade(90))
print(student.add_grade(80))
print(student.add_grade(70))
print(student.get_average())
```
