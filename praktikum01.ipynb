```python
class AkunBank:
    def __init__(self, saldo):
        self._saldo = saldo
    def cek_saldo(self):
        return self._saldo
    def setor(self, jumlah):
        if jumlah > 0:
            self._saldo += jumlah

rekening = AkunBank(1000)
rekening.setor(500)
print(rekening.cek_saldo())  # Output: 1500
rekeningA = AkunBank(1500)
rekeningA.setor(1000)
print(rekeningA.cek_saldo()) 
 # Output: 1500
```
